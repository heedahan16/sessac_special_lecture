### 데이터 분석가(Data Analysis, DA) - Data Analysis Platform
- Log 정의
- 데이터 수집
  - 로그 모니터링
- 데이터 분석
  - 실험 결과 분석: AB Test.
  - 퍼널 분석
  - ctr, cvr 확인
- 인사이트 도출
  - 대시보드 제작
  - 방향성 제시

#### Log Data
사용자가 해당 앱을 어떻게 이용하는지 분석하기 위해 수집하는 데이터.
- 이러한 데이터를 정의한 다음 해당 로그를 분석하기 위해 여러가지 테스트를 진행한다.
  - A/B test
  - Funnel 분석 등.

### 시각화 대시보드
인사이트 도출을 위해 데이터를 여러가지 차트로 제작하여 표현한 것
- 예시: 리텐션 차트(코호트 분석)
  - 같은 기간이 지난 후 cohort 간 리텐션 비교.
  - 같은 시간대 cohort 간 비교.

- 공공데이터를 이용해서 차트 만들어보는 포트폴리오 유리.

#### 데이터 분석을 하기 위해 필요한 전체 Flow
예시) 'A' 서비스가 새롭게 출시되어 해당 서비스에 대한 사용자들의 반응을 파악하고자 함.
- 서비스 진입 후 동의 완료까지 하는 수치 파악.
1. 해당 화면 컴포넌트 별 로그 정의
  - top_banner: :click 로그
2. 개발자 로그 작업
  - top_banner::click 로그를 code에 적용
3. 로그 인입 테스트
  - 앱에서 top_banner 클릭 후 해당 로그가 수집되는지 live 체크
4. 데이터 추출
  - top_banner에 대한 count값 추출
5. 의사결정에 필요한 차트 생성.
  - top_banner에 대한 CTR 수치 차트로 확인

### 분석 직무를 하기 위해서 어떤 스킬을 쌓아야 하는가?
> #### 신입 직군
- 채용 공고 확인.
  - 주요업무
  - 우대사항
  - 신입의 경우 위에서 언급한 경력 또는 기술을 보유하지 않았더라도 DB나 개발 관련 기초 지식이 있고 관련 분야 일에 열정과 관심이 있다면 지원 가능함.
    - 기초 지식과 열정은 포트폴리오로 판단.
    - 회사에서 원하는 업무를 해봤는지를 파악.

> #### 신입 + 경력 직군
- 전략가의 업무
- SQL을 사용해 직접 데이터를 추출
- 데이터 시각화(Tableau, PowerBI)등에 대한 경험.
  - 기업에서 많이 사용하는 프로그램.
    - 토스에서 가산점.
- 데이터 분석 방법론 및 관련 내용에 대한 높은 이해도
  - Retention
  - Cohort
  - Funnel 등

### 포트폴리오는 어떻게 준비하면 좋을까?
- 분석 리포트 작성
  1. 본인의 역량을 보여줄 수 있는 분석 리포트 작성.
    - 리포트는 Word, Notion, GitHub 등 본인이 편한 툴을 활용하여 정리하는 것을 추천.
      - 기업에 제출할 때는 가급적 PDF 형태로 제출하는 것이 파일 내용이 손상되지 않고 좋음.
    - 공공데이터 및 민간데이터를 활용하여 인사이트를 보여줄 수 있는 리포트 작성
      - ERD 작성
    - 온라인 커머스 분석 예시
      - 온라인 커머스 sample 데이터에서 필요한 내용에 대한 Mart Table을 만든 후 데이터 추출 및 대시보드 제작
        - Sample  데이터 ERD(한 눈에 테이블들의 연관관계 파악)
          - 카테고리 별 테이블 분리
          - 키값을 기준으로 하나의 테이블로 병합
        - Mart Table 생성(원본 테이블)
          - 원하는 지표 추출
        - 대시보드 생성
          - 요약본
          - 지도, 카테고리 값 등을 하나의 리포트로 요약.
          - 태블로 사이트 활용.
            - 아이디 만들고 링크 제출해서 포트폴리오로 사용 가능.
- 분석 or 시각화 공모전 참여
  2. 데이터 관련 공모전 참여
    - 공모전은 데이터분석 or 시각화 아이디어 관련 대회를 참여하는 것이 좋음.
      - 예시) 빅데이터 기반 시각화 공모전 참여 내용.
        - 대용량의 데이터를 전처리 후 화면으로 결과값 출력.

### HOW? (대시보드 제작시 아이디어 도출 방법)
사용하고자 하는 데이터를 먼저 파악한 뒤 궁금한 내용들과 이를 뒷받침 할 수 있는 추가 데이터를 찾아서 결론을 도출
1. 데이터 탐색
  - 여러 가입경로로 온라인 커머스에 가입.
  - 유의미한 가입경로를 파악하고자 함.
  - 경로별 진입 고객값을 다 뽑아봄.
    - 가장 많은 양을 도출한 경로에 광고 노출을 많이 하겠다!
2. 데이터 전처리
3. 차트 분석
4. 결론 도출


### 포트폴리오 구성 꿀팁!
- 자기소개서 + 포트폴리오
  - 신입 직군 지원 시 두가지는 필수 요소!
    - Word형식의 포트폴리오의 경우 간단하게 요약된 정보로 전달하는 것이 중요.
      - 핵심만 추려서 요약된 내용을 보여줘야 함 -> 눈대중으로 확인.
    - 과제의 목표와 사용된 기술이 주로 서술되어야 함.
    - 결과적으로 만들어낸 impact 에 대해서 결과물도 같이 첨부하면 좋음.
      - 회사에서는 프로세스보다 결과가 중요.
    - 공공 데이터 or 본인이 수집한 데이터를 가지고 분석해본 결과물을 포트폴리오에 첨부해도 좋음.
    - 잘 다룰 수 있는 본인만의 tool이 있다면 그걸 보여줄 수 있도록 시나리오를 설계해서 포트폴리오를 위한 과제를 해보는 것도 도움이 됨.
      - 구글 사이트에 샘플데이터를 만들 수 있는 사이트가 있음.
        - 데이터 만들기 + 분석 까지 가능.
        - 열정 넘치는 모습도 보여줄 수 있음.

### if 내가 취준생으로 돌아간다면 준비했을 것들.
1. 가고자 하는 기업의 JD(직무요건)를 유심히 살펴보고 기술 spec을 쌓을 것.
    - 기업들이 필요로 하는 열걍에 대한 trend를 알 수 있음.
2. 원하는 기업이 없다 해도 원하는 직무에 대해선 여러 기업을 조사해보는 것도 필요.
    - 해당 분야에 자세한 역량 선호도를 알 수 있음.
3. 본인의 이력서를 잘 활용할 수 있도록 미리 링크드인 같은 비즈니스 소셜 플랫폼을 시작해보는 것도 추천
    - 해당 플랫폼에서 여러 업계 소식부터 다양한 채용 공고까지 살펴볼 수 있으니 잘 활용해보기!
    - 이직할 때 많이 활용하지만 신입일 때부터 쓰면 인맥까지 쌓을 수 있음.
        - 원하는 기업을 다니고 있는 사람의 리스트가 나옴.
        - 프로필에 링크가 나와 있음.
        - 친구 신청을 해서 나를 어필하기.
            - 메시지에 간단하게 남겨봐도 좋음.
            - 실무자 면접에 대비할 수 있음.
